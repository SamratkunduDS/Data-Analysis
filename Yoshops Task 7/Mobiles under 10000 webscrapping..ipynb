```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
Product_name =[]
Prices = []
Description=[]
Reviews=[]
```

```python
for i in range(2,10):
    url = "https://www.flipkart.com/search?q=mobile+under+10000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div",class_ = "_1YokD2 _3Mn1Gg")
    np = soup.find("a",class_ = "_1LKTO3").get("href")
    cnp = "https://www.flipkart.com"+np
    names = box.find_all("div",class_ = "_4rR01T")
    for i in names:
        name = i.text
        Product_name.append(name)
    prices = box.find_all("div",class_ = "_30jeq3 _1_WHN1")
    for i in prices:
        price = i.text
        Prices.append(price)
    desc = box.find_all("ul",class_ = "_1xgFaf")
    for i in desc:
        desc = i.text
        Description.append(desc)
    reviews = box.find_all("div",class_ = "_3LWZlK")
    for i in reviews:
        reviews = i.text
        Reviews.append(reviews)
```

```python
print(len(Product_name))
print(len(Prices))
print(len(Description))
print(len(Reviews))
```
