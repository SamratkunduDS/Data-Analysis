```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
Product_name =[]
Prices = []
Description=[]
Reviews=[]
```

```python
for i in range(2,10):
    url = "https://www.flipkart.com/search?q=laptops+under+20000&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_18_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_18_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=laptops+under+20000%7CLaptops&requestId=faa6291a-cd64-49bd-871c-1ffc9dc3b508&as-searchtext=laptops+under+2000&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div",class_ = "_1YokD2 _3Mn1Gg")
    np = soup.find("a",class_ = "_1LKTO3").get("href")
    cnp = "https://www.flipkart.com"+np
    names = box.find_all("div",class_ = "_4rR01T")
    for i in names:
        name = i.text
        Product_name.append(name)
    prices = box.find_all("div",class_ = "_30jeq3 _1_WHN1")
    for i in prices:
        price = i.text
        Prices.append(price)
    desc = box.find_all("ul",class_ = "_1xgFaf")
    for i in desc:
        desc = i.text
        Description.append(desc)
    reviews = box.find_all("div",class_ = "_3LWZlK")
    for i in reviews:
        reviews = i.text
        Reviews.append(reviews)
```

```python
print(len(Product_name))
print(len(Prices))
print(len(Description))
print(len(Reviews))
```

```python
df = pd.DataFrame({"Prices":Prices,"Description":Description})
df.to_csv("LAPTOP.csv")
```

```python
df = pd.DataFrame({"Product Name":Product_name})
df.to_csv("prodname.csv")
```

```python
df = pd.DataFrame({"Reviews":Reviews})
```

```python
df.to_csv("Reviw.csv")
```
